```python
house_data=house_data.drop_duplicates()
house_data
```

```python
house_data=house_data.drop(columns=['Pid'])
house_data
```

```python
house_data['Price']=house_data['Price'].str.replace(r'\D','',regex=True)
house_data['Price']=pd.to_numeric(house_data['Price'],errors='coerce')
house_data
```

```python
def get_neighborhood(Address):
    parts=Address.lower().split()
    if 'Lagos' in parts:
         index=parts.index('Lagos')
         if index>0:
             return parts[index-1]
         return parts[-1]
house_data['Neighborhood']=house_data['Address'].apply(get_neighborhood)
house_data
```
